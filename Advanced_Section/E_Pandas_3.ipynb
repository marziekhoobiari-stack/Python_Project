```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
new_df = df.copy()

new_df = new_df.dropna(subset='salary_year_avg')

new_df.head(5)
```

```python
frac_df = df.sample(frac=0.1)

frac_df.head(5)
```

```python
cleaned_df = df.copy()

median_salary = cleaned_df['salary_year_avg'].median()
cleaned_df['salary_year_avg'] = cleaned_df['salary_year_avg'].fillna(median_salary)

cleaned_df.head(5)
```
