```python
import ast
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

df
```

```python
target_job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']

df_filtered = df[df['job_title_short'].isin(target_job_titles)]

df_filtered
```

```python
df_filtered = df_filtered[df_filtered['job_country']=='United States']

df_filtered
```

```python
schedule_type = ['Full-time','Contractor']

df_filtered = df_filtered[df_filtered['job_schedule_type'].isin(schedule_type)]

df_filtered
```

```python
df_filtered = df_filtered.dropna(subset='salary_year_avg')
df_filtered
```

```python
df_filtered['job_title_type'] = df_filtered['job_title_short'] + ' - ' + df_filtered['job_schedule_type']

df_filtered
```

```python
df_grouped = df_filtered.groupby('job_title_type').agg(
    posting_count=('job_title_short','count'),
    avg_salary=('salary_year_avg','mean')

)
df_grouped
```

```python
df_grouped = df_grouped.sort_values(by='posting_count',ascending=False)
df_grouped
```

```python
df_grouped.plot(kind='scatter',x='posting_count',y='avg_salary')
plt.xlabel('posting count')
plt.ylabel('average yearly salary')
plt.legend()

```

```python
df_grouped = df_grouped.reset_index()
for job_title_type in df_grouped['job_title_type'].unique():
   subset = df_grouped[df_grouped['job_title_type'] == job_title_type]
   plt.scatter(subset['posting_count'], subset['avg_salary'], label=job_title_type)

plt.xlabel('Number of Jobs')
plt.ylabel('Average Yearly Salary (USD)')
plt.title('Average Yearly Salary vs. Job Count for Top Roles by Schedule Type')
plt.legend()
plt.show()
```
