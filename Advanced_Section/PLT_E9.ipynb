```python
import ast
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DS = df[df['job_title_short']=='Data Scientist'].copy()
```

```python
df_DS = df['salary_hour_avg'].plot(kind='hist', bins=40, color='blue', edgecolor='black')
plt.xlabel('Hourly Salary')
plt.ylabel('Number of Jobs')
```

```python
df_skills = df.dropna(subset=['job_skills']).copy()
```

```python
df_skills['skill_count'] = df_skills['job_skills'].apply(len)
df_skills['skill_count'].plot(
    kind='hist',
    bins=55,
    color='magenta',
    edgecolor='black'
)

plt.title('Distribution of Job Postings by Number of Skills Listed')
plt.xlabel('Number of Skills')
plt.ylabel('Number of Job Postings')
```

```python
df_DE = df[df['job_title_short'] == 'Data Engineer'].copy()
df_DE
```

```python
df_DE['salary_year_avg'].plot(kind='hist', bins=50, color='red', edgecolor='black')
median_salary = df_DE['salary_year_avg'].median()
average_salary = df_DE['salary_year_avg'].mean()
plt.axvline(
    median_salary,
    color='blue',
    linestyle='dashed',
    label=f'Median: ${median_salary:,.0f}'
)
plt.axvline(
    average_salary,
    color='green',
    linestyle='dashed',
    label=f'Mean: ${average_salary:,.0f}'
)
plt.legend()
plt.xlabel('Yearly Salary')
plt.ylabel('Number of Jobs')
```
