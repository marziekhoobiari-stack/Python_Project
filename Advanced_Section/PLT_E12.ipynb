```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
countries = ['United States', 'Canada']

df_filtered = df[(df['job_country'].isin(countries)) & (df['job_title_short']=='Data Analyst')].dropna(subset=['salary_hour_avg'])
```

```python
country_data = [df_filtered[df_filtered['job_country']== country]['salary_hour_avg'] for country in countries]
```

```python
import matplotlib.pyplot as plt
plt.boxplot(country_data, labels=countries)
plt.xlabel('')
plt.ylabel('Hourly Salary ($USD)')
plt.ylim(0,150)
plt.title('Comparison of Hourly Salaries for Data Analysts in the US and Canada')
plt.show()
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_fil = df[df['job_title_short'].isin(job_titles)].dropna(subset='salary_hour_avg')

job_list = [df_fil[df_fil['job_title_short'] == title]['salary_hour_avg'] for title in job_titles]
```

```python
plt.boxplot(job_list, labels=job_titles)
plt.xlabel('Job Title')
plt.ylabel('Hourly Salary (USD)')
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'${y}'))

```
