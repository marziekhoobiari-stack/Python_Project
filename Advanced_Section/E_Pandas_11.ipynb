```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from datasets import load_dataset
```

```python
dataset  =load_dataset('lukebarousse/data_jobs')
```

```python
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
import ast
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

df
```

```python
df_exploded = df.explode('job_skills')
```

```python
df_exploded
```

```python
df_exploded_CE = df_exploded[df_exploded['job_title_short'] == 'Cloud Engineer'].copy()

df_exploded_CE
```

```python
skills_count = df_exploded_CE['job_skills'].value_counts().head(10)

skills_count
```

```python
skills_count.plot(kind='bar')
plt.xlabel('Skills')
plt.ylabel('Count')
plt.title('Top 10 most common skills for cloud engineers')
plt.show()
```
