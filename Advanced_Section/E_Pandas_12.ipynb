```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
```

```python
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
import ast
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_exploded = df.explode('job_skills')

df_exploded
```

```python
df_exploded_grouped = df_exploded.groupby('job_skills')['salary_year_avg'].mean()

df_exploded_grouped = df_exploded_grouped.reset_index(name='average_salary')
```

```python
df_exploded_grouped = df_exploded_grouped.sort_values(by='average_salary', ascending=False)
df_exploded_grouped
```

```python
df_exploded_grouped.head().plot(kind='bar', x='job_skills', y='average_salary')
```

```python
df_gr = df_exploded.groupby('job_title_short')['job_skills'].nunique()

df_gr = df_gr.reset_index(name='num_unique_skills')

df_gr.head()
```
