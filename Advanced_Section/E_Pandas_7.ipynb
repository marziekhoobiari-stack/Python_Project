```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from datasets import load_dataset
```

```python
dataset = load_dataset('lukebarousse/data_jobs')
```

```python
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.to_period('M')

df
```

```python
df_us_jobs = df[df['job_country'] == 'United States'].copy()
```

```python
df_monthly_postings = df_us_jobs.groupby('job_posted_month').size().reset_index(name='postings_count')
```

```python
merged_df = pd.merge(df_us_jobs, df_monthly_postings, on='job_posted_month', how='right')
merged_df.head()
```

```python
df_not_remote_jobs = df[df['job_work_from_home'] == False].copy()
```

```python
df_not_remote_salary = df_not_remote_jobs.groupby('job_title_short')['salary_year_avg'].mean().reset_index()
df_not_remote_salary.rename(columns={'salary_year_avg': 'salary_not_remote_avg'}, inplace=True)

df_not_remote_salary
```

```python
df_not_remote_jobs
```

```python
df_not_remote_jobs.merge(df_not_remote_salary, on='job_title_short', how='right')
```

```python
df_company_salary = df.groupby('company_name')['salary_year_avg'].mean().reset_index()
df_company_salary 
```

```python
df_company_remote = df['company_name'].value_counts().reset_index()
df_company_remote
```

```python
df_company_remote = df_company_remote.rename(columns={'count':'remote count'})
df_company_remote
```

```python
df_company_salary.merge(df_company_remote, on='company_name')
```
