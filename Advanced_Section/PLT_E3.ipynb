```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import ast
```

```python
from datasets import load_dataset

```

```python
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

df
```

```python
job_roles = ['Senior Data Analyst', 'Senior Data Scientist', 'Senior Data Engineer']

fig, ax = plt.subplots(3, 1)

for i, role in enumerate(job_roles):
    df_role = df[df['job_title_short'] == role]
    exploded_skills = df_role['job_skills'].explode()
    job_skills = exploded_skills.value_counts().head(3)

    ax[i].barh(job_skills.index, job_skills.values)
    ax[i].set_title(role)

plt.tight_layout()
plt.show()



```
