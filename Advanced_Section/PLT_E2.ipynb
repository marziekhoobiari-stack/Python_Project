```python
import pandas as pd
```

```python
import ast
```

```python
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
```

```python
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
schedules = ['Full-time', 'Part-time', 'Contractor', 'Internship']
```

```python
df
```

```python
df_grouped = df.groupby(['job_title_short', 'job_schedule_type']).size().reset_index(name='count')

df_grouped 
```

```python
fig , ax = plt.subplots(2,2)
ax = ax.flatten()

for i , schedule in enumerate(schedules):
    top_3 = df_grouped[df_grouped['job_schedule_type']==schedule].sort_values(by='count',ascending=False).head(3)
    ax[i].bar(top_3['job_title_short'], top_3['count'])
    ax[i].tick_params(axis='x', labelrotation=45)

    for label in ax[i].get_xticklabels():
        label.set_horizontalalignment('right')


fig.tight_layout()
```
