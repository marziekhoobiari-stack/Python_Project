```python
from datasets import load_dataset
```

```python
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.to_csv('data_jobs_export.csv', index=False)
```

```python
data_scientist_jobs = df[df['job_title_short']=='Data Scientist'].copy()
```

```python
data_scientist_jobs.to_csv('data_scientist_jobs.csv', index=False)
```

```python
new_york_jobs = df[df['job_location']=='New York'].copy()
```
