```python
import ast
import pandas as pd
import matplotlib.pyplot
```

```python
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_ce = df[df['job_title_short']=='Cloud Engineer'].copy()

df_ce
```

```python
df_ce['month_no'] = df_ce['job_posted_date'].dt.month
df_ce
```

```python
df_ce = df_ce.sort_values(by='month_no')
df_ce
```

```python
df_ce['month_name'] = df_ce['job_posted_date'].dt.month_name()
df_ce
```

```python
df_grouped = df_ce.groupby(['month_no','month_name']).agg(job_count=('month_name','count')).reset_index().sort_values('month_no')
df_grouped
```

```python
import matplotlib.pyplot as plt
df_grouped.plot(x='month_name',y='job_count',
    kind='line',
    linewidth=4,
    linestyle=':',
    color='purple',
    marker='o',
    figsize=(12,5)
)
plt.title('Job Postings Trend for Cloud Engineers')
plt.ylabel('Count of Job Postings')
plt.xlabel('')
plt.xticks(ticks=range(len(df_grouped)), labels=df_grouped['month_name'])
plt.legend().set_visible(False)
plt.show()
```

```python
df_fil_gr = df_fil.groupby('company_name').agg(count=('company_name','count')).sort_values(by='count',ascending=False).head(5)
df_fil_gr
```

```python
colors=['purple','blue','green','orange','red']
plt.bar(df_fil_gr.index,df_fil_gr['count'], color=colors)
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.ylabel('Number of Job Postings')
plt.title('Top 5 Companies hiring Senior Date Analysts by Number of Postings', fontsize=14)
plt.show()
```
