```python
import pandas as pd
```

```python
from datasets import load_dataset
```

```python
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_date_str'] = df['job_posted_date'].apply(lambda time:time.strftime('%Y-%m-%d'))
```

```python
df['job_posted_date_str']
```

```python
import datetime
datetime.datetime.now()
```

```python
df['days_since_posted'] = df['job_posted_date'].apply(lambda time: (datetime.datetime.now()-time).days)
```

```python
df['days_since_posted']
```

```python
df_filtered = df['salary_year_avg'].dropna().copy()

df_filtered = df_filtered.to_frame()  # convert to DataFrame
```

```python
df_filtered
```

```python
df_filtered['salary_category'] = df_filtered['salary_year_avg'].apply(
    lambda s: 'Low' if s < 60000 else 'Medium' if s < 100000 else 'High'
)

df_filtered[['salary_year_avg', 'salary_category']]
```
